```python
import pandas as pd
df = pd.read_csv("session4_dummy_students_data.csv.xls")
```

```python
df['average'] = df[['math', 'science', 'english']].mean(axis=1)
display(df.head())
```

```python
df['result'] = df['average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
display(df.head())
```

```python
class_average = df['average'].mean()
print(f"Class Average: {class_average:.2f}")
```

```python
highest_average = df['average'].max()
lowest_average = df['average'].min()

print(f"Highest Average: {highest_average:.2f}")
print(f"Lowest Average: {lowest_average:.2f}")
```

```python
highest_average_score = df['average'].max()
top_students = df[df['average'] == highest_average_score][['name', 'average']]

print("Top student(s):")
display(top_students)
```

```python
students_low_attendance = df[df['attendance_percent'] < 75][['name', 'attendance_percent']]

print("Students with attendance below 75:")
display(students_low_attendance)
```

```python
students_few_assignments = df[df['assignments_submitted'] < 6][['name', 'assignments_submitted']]

print("Students who submitted fewer than 6 assignments:")
display(students_few_assignments)
```

```python
df.to_csv('session4_student_report.csv', index=False)
print("Final report saved to 'session4_student_report.csv'")
```
